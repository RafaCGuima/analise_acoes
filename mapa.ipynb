```python
import pandas as pd
import matplotlib
import yfinance as yf
```

```python
df_ibov = pd.read_csv('IBOV.csv', sep=';', skiprows=1,)
df_ibov
```

```python
df_ibov.index[:-2]
```

```python
ativos_ibov = list(df_ibov.index[:-2])
ativos_ibov
```

```python
ativos_ibov = map(lambda a: f"{a}.SA", ativos_ibov)
ativos_ibov_sa = " ".join(list(ativos_ibov))
ativos_ibov_sa
```

```python
yf.pdr_override()
```

```python
cotacoes_ativos = yf.download(tickers=ativos_ibov_sa, period='6mo')
```

```python
cotacoes_ativos['Adj Close'].info()
```

```python
cotacoes_ativos_fechamento = cotacoes_ativos['Adj Close']
cotacoes_ativos_fechamento
```

```python
retorno = cotacoes_ativos_fechamento.pct_change()
retorno_acumulado = (1 + retorno).cumprod()
retorno_acumulado.iloc[0] = 1

retorno_acumulado
```

```python
percentual_do_periodo = retorno_acumulado.iloc[-1]
ordenados = percentual_do_periodo.sort_values()
dez_melhores = ordenados[-11:-1]
dez_melhores
```

```python
dez_melhores_porcento = (dez_melhores -1)* 100
dez_melhores_porcento
```

```python
dez_melhores_porcento.plot.bar(figsize=(20,5))
```
